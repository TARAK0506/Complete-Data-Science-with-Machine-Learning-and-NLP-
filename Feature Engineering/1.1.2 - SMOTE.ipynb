# SMOTE
Synthetic Minority Over-sampling Technique (SMOTE) is a popular algorithm to generate synthetic samples for the minority class. It works by selecting two or more similar instances (using a distance metric) and generating a new instance that is a convex combination of the selected instances. This technique is particularly useful when the minority class is under-represented in the dataset.

```python
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_samples=1000, n_features=2, n_redundant=0, n_repeated=0,  n_clusters_per_class=1, weights=[0.90], random_state=12)
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(X, columns=['feature1', 'feature2'])
df['target'] = y
```

```python
df.to_csv('data.csv', index=False)
```

```python
df.head()
```

```python
df['target'].value_counts()
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['feature1'], df['feature2'], c=df['target'])
plt.show()
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(sampling_strategy='minority')
X_sm, y_sm = smote.fit_resample(X, y)

df_sm = pd.DataFrame(X_sm, columns=['feature1', 'feature2'])
df_sm['target'] = y_sm

df_sm['target'].value_counts()
```

```python
plt.scatter(df_sm['feature1'], df_sm['feature2'], c=df_sm['target'])
plt.show()
```

```python
df_sm.to_csv('data_smote.csv', index=False)
```
