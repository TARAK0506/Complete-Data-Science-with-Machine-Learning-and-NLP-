```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.plot(x, y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Simple plot')
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.plot(x, y, color = 'red', linestyle = 'dashed', linewidth = 2, marker = 'o', markerfacecolor = 'blue', markersize = 12)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Custumized plot')
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.plot(x, y, 'ro-')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Simple plot')
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.plot(x, y, 'ro-')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Simple plot')
plt.grid()
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.plot(x, y, 'ro-')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Simple plot')
plt.grid(axis = 'y')
plt.show()
```

```python
# Multiple plots

x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [1, 2, 3, 4, 5]

plt.figure(figsize=(10, 5))
plt.subplot(1,2,1)
plt.plot(x, y1,color = 'red')
plt.title('First plot')

plt.subplot(1,2,2)
plt.plot(x, y2, color = 'blue')
plt.title('Second plot')

plt.suptitle('Two plots')
plt.show()
```

```python
# Multiple plots

x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [1, 2, 3, 4, 5]

plt.figure(figsize=(10, 5))
plt.subplot(2,1,1)
plt.plot(x, y1,color = 'red')
plt.title('First plot')

plt.subplot(2,1,2)
plt.plot(x, y2, color = 'blue')
plt.title('Second plot')

plt.suptitle('Two plots')
plt.show()
```

```python
# Bar plot

categories = ['A', 'B', 'C', 'D', 'E']
values = [7, 3, 12, 5, 18]

plt.bar(categories, values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Simple bar plot')
plt.show()
```

```python
# Histograms

import numpy as np

data = np.random.randn(1000)

plt.hist(data, bins = 30, color = 'blue', alpha = 0.5)
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

```python
# Scatter plot

x = np.random.randn(1000)
y = np.random.randn(1000)

plt.scatter(x, y, color = 'red',marker='*')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Scatter plot')
plt.show()
```

```python
# Pie chart

sizes = [3, 4, 6, 2]
labels = ['A', 'B', 'C', 'D']

plt.pie(sizes, labels = labels, autopct = '%1.1f%%')
plt.title('Simple pie chart')
plt.show()
```

```python
# Box plot

data = np.random.randn(100)

plt.boxplot(data)
plt.title('Box plot')
plt.show()
```

```python
# Violin plot

data = [np.random.randn(100) for _ in range(4)]

plt.violinplot(data)
plt.title('Violin plot')
plt.show()
```
